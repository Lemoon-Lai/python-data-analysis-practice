## Conda基础命令
``` python
# 检查conda版本
conda --version

# 查看已安装的环境
conda info -e

# 升级conda
conda update conda

# 帮助
conda update --help
```

## 管理Conda环境
```python
# 管理conda环境
conda env --help

# 当前environments 信息 list env
conda info --ENVS

# 安装一个名为py35的环境，指定Python版本是3.5（不用管是3.5.x，conda会为我们自动寻找3.4.x中的最新版本）
conda create --name py35 python=3.5
conda create -n py35 python=3.5

# 安装好后，使用activate激活某个环境
activate py35    # for Windows
source activate py35    # for Linux & Mac

# 此时，再次输入
python --version

# 如果想返回默认的python 2.7环境，运行
deactivate py35 # for Windows
source deactivate py35 # for Linux & Mac
 
# 删除一个已有的环境
conda remove --name py35 --all
```

## Conda的包管理
```python
# 查看已经安装的packages
conda list

# 在ENV环境下，安装名为PACKAGENAME的包
# (如果不用-n指定环境名称，则被安装在当前活跃环境
#  也可以通过-c指定通过某个channel安装)
conda install -n ENVS PACKAGENAME
pip install boltons

# 更新名为PACKAGENAME的包
conda update -n ENVS PACKAGENAME

# 删除package
conda remove -n ENVS PACKAGENAME

# 搜索包PACKAGENAME
conda search PACKAGENAME

## clone env
conda create --clone py35 --name py35-2
```

## 管理Python 
```python
# set python version
conda create -n yourname python=x.x packages-your-want

# manage python，当前python 版本  
conda search --full-name python
conda create --name snakes python=3
python --version
```

## share env
```python
source activate yourname
activate yourname
conda env export > envrionment.yml
conda env create -f environment.yml
#another way
conda list -e>spec-file.txt
conda create --name myname --file spec-file.txt
```
